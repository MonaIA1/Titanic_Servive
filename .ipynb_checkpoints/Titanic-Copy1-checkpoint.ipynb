The aim of this project is to identify the survived passengers and their characterisic (invistigation of the context of the survived passengers). ### Filling up missing values

```python
# fill nulls in Age column

grouped_tr = titanic_tr.groupby(['Sex','Pclass', 'Title']).mean().to_dict()["Age"]
grouped_ts = titanic_ts.groupby(['Sex','Pclass', 'Title']).mean().to_dict()["Age"]

#fill nulls in train subset
titanic_tr.loc[titanic_tr.Age.isna(), "Age"] = titanic_tr.loc[titanic_tr.Age.isna()].apply(lambda x: 
                                                                    grouped_tr[x["Sex"], x["Pclass"], x["Title"]] , axis=1)

#fill nulls in test subset
titanic_ts.loc[titanic_ts.Age.isna(), "Age"] = titanic_ts.loc[titanic_ts.Age.isna()].apply(lambda x: 
                                                                    grouped_ts[x["Sex"], x["Pclass"], x["Title"]] , axis=1)
```

```python
# check values in the table to make sure of accurate filling
titanic_tr.Age.head(10)
titanic_ts.Age.head(10)
```

```python
# convert Survived back to int with round
titanic_tr.Age = round(titanic_tr['Age']).astype(int)
titanic_tr.head()
titanic_ts.Age = round(titanic_ts['Age']).astype(int)
titanic_ts.head()
```

## Modeling

### Predicting Survival using Random Forest Tree (RT)

```python
# pridect null values using RT algorithm

# set X, y for RT
X = titanic_tr.drop(['Survived', 'PassengerId','Ticket', 'Name'], axis = 1)
y = titanic_tr['Survived']

# Do train test split with 20% as test data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle  = True, random_state  = 42, stratify = y )

# Do Scaling for data

ss = StandardScaler()
X_train = ss.fit_transform(X_train)
X_test = ss.transform(X_test)
```

```python
# baseline model
# pick the max value of each class as a baseline
y.value_counts()

baseline = max (y.value_counts()/y.shape[0])
baseline
```

```python
#check best parameters

rt = RandomForestClassifier()

search_grid = {'n_estimators':[500,1000,2000], 'max_depth': [4,10, 50, 100],'random_state':[1]}
model = GridSearchCV(estimator= rt, param_grid = search_grid, scoring = 'neg_mean_squared_error', n_jobs = 1)
model.fit(X_train,y_train)
model.best_params_
```

```python
# initialize the RT model
forest = RandomForestClassifier(n_estimators = 500, max_features ='sqrt', random_state= 1, max_depth=10, oob_score = True)
forest.fit(X_train, y_train)
y_pred = forest.predict(X_test)

# check model scorse
print ('train score ', forest.score(X_train, y_train))
print ('test score ', forest.score(X_test, y_test))

# check accuracy score
print('\nmodel accuracy score: ', accuracy_score(y_test ,y_pred))

# confusion matrix
print('\nConfusion matrix:\n', confusion_matrix(y_test, y_pred))

#classification_report
print('\nclassification report:\n', classification_report(y_test, y_pred))
```

```python
# plot feature importance
f_ = pd.Series(forest.feature_importances_, X.columns).sort_values(ascending=False)
f_
# f_.plot(kind = 'barh')
```

```python
# pridect null values using RT algorithm

# set X, y for RT                        
       
X = titanic_tr[['Fare', 'Age',  'Pclass', 'Title_Mr', 'Sex_male',  'Sex_female']]
y = titanic_tr['Survived']
 
# Do train test split with 20% as test data

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle  = True, random_state  = 42)

# Do Scaling for data

ss = StandardScaler()
X_train = ss.fit_transform(X_train)
X_test = ss.transform(X_test)

# initialize the RT model
forest1 = RandomForestClassifier(n_estimators = 500, max_features ='sqrt', random_state= 1, max_depth=10, oob_score = True)
forest1.fit(X_train, y_train)
y_pred = forest1.predict(X_test)

# check scores
print ('train score ', forest1.score(X_train, y_train))
print ('test score ', forest1.score(X_test, y_test))

# check accuracy score
print('\nmodel accuracy score: ', accuracy_score(y_test ,y_pred))

# confusion matrix
print('\nConfusion matrix:\n', confusion_matrix(y_test, y_pred))

#classification_report
print('\nclassification report:\n', classification_report(y_test, y_pred))
```

### Predicting Survival using Supervised Victor Machein (SVM)

```python
#create intersection list of both headers(train, and test subsets)

temp = set(list(titanic_tr.columns)) 
X_features = [value for value in (list(titanic_ts.columns)) if value in temp] 
     
X_features.remove ('PassengerId')
X_features.remove ('Ticket')
X_features.remove ('Name')
X_features.remove ('Survived')

```

```python
# set X, y for SVC
X =titanic_tr[X_features]
y = titanic_tr['Survived']

# Do train test split with 20% as test data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle  = True, random_state  = 42, stratify = y )

# convert Survived back to int
titanic_tr.Survived = titanic_tr.Survived.astype(int)

# Do Scaling for data

ss = StandardScaler()
X_train = ss.fit_transform(X_train)
X_test = ss.transform(X_test)
```

```python
# baseline model
# pick the max value of each class as a baseline
y.value_counts()

baseline = max (y.value_counts()/y.shape[0])
baseline
```

```python
# initialize "Support Vector Classifier" 

clf = SVC() 

# fitting x samples and y classes 
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)

# check scores
print ('train score ', clf.score(X_train, y_train))
print ('test score ', clf.score(X_test, y_test))

# check accuracy score
print('\nmodel accuracy score: ', accuracy_score(y_test ,y_pred))

# confusion matrix
print('\nConfusion matrix:\n', confusion_matrix(y_test, y_pred))

#classification_report
print('\nclassification report:\n', classification_report(y_test, y_pred))
```

## Predicting Survaival using KNN

```python
# set X, y for KNN

X = titanic_tr[X_features]
y = titanic_tr['Survived']

# Do train test split with 20% as test data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle= True, random_state= 42,
                                                    stratify = y )

# Do Scaling for data

ss = StandardScaler()
X_train = ss.fit_transform(X_train)
X_test = ss.transform(X_test)
```

```python
# baseline model
# pick the max value of each class as a baseline
y.value_counts()

baseline = max (y.value_counts()/y.shape[0])
baseline
```

```python
# check best parameter

params = {'n_neighbors':[2,5,10,15,20]}

knn = neighbors.KNeighborsClassifier()

model = GridSearchCV(knn, params, cv=3)
model.fit(X_train,y_train)
model.best_params_
```

```python
#from sklearn.model_selection import StratifiedKFold


# Intialize our model
knn = KNeighborsClassifier(n_neighbors=15)


# fitting x samples and y classes 
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)

# use stratifiedkfold in order to balance the classes in the stratified folds
folds = StratifiedKFold(n_splits=3, shuffle=True, random_state = 5)

#train model with cv = flods 
cv_scores = cross_val_score(knn, X_train, y_train, cv= folds)



#  Score our model
print('cross val mean score ', cv_scores.mean())


# Score our model
print('train score ', knn.score(X_train, y_train))
print('test score ', knn.score(X_test, y_test))

print('\n', '--------')

#print each cv score (accuracy) and average them
print('cv_scores',cv_scores)
print('cv_scores mean:{}'.format(np.mean(cv_scores)))


# check rmse
print('\nmodel accuracy score: ', accuracy_score(y_test ,y_pred))

# confusion matrix
print('\nConfusion matrix:\n', confusion_matrix(y_test, y_pred))

#classification_report
print('\nclassification report:\n', classification_report(y_test, y_pred))
```
